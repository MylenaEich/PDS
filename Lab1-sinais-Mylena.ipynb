## 1) Gerar os seguintes sinais e visualiza-los (comandos plot e stem). 

```python
from pylab import *
import numpy as np
```

### a) impulso unitário, para valores de n (abscissa) entre -10 e 10 (stem), inclusive.

```python
x = arange(-10,11)
y = zeros(len(x))

origem = int((len(x)-1) / 2)
y[origem] = 1

plt.stem(x, y, use_line_collection=True)
plt.show()
```

### b) degrau unitário, para n entre -10 e 10 (stem), inclusive.


```python
x = arange(-10, 11)
y = zeros(len(x))

origem = int((len(x) - 1) / 2)

posicao_inicial_impulso = origem
for i in range(origem + 1):
    y[i+origem] = 1

plt.stem(x, y, use_line_collection=True)
plt.show()
```

### c) ruido branco gaussiano (média zero, variância unitária) com 40 valores, para n=0 a 39 (stem).

```python
n = 40
x = range(0, n)
np.random.seed(0)
ruido = np.random.randn(n)

plt.stem(x, ruido)
plt.show()
```

### h) senoide de 60 Hz, (amostragem = 800 amostras/s), com 70 valores, para n de 0 a 69 (plot).

### l) exponencial decrescente para n = 0 a 60 (stem).

### m) exponencial oscilante decrescente $x[n]=z^n$, para n=0 a 90, parte real (stem).

```python
n = 90
periodo = 15
frequencia = 1 / periodo
omega = 2 * pi * frequencia
sigma = -0.042

x = arange(0, n+2)

complex_num = exp(complex(sigma, omega) * x*0.7)

complexo_real = [num.real for num in complex_num]

plt.stem(x, complexo_real, use_line_collection=True)
plt.show()
```

### n) exponencial oscilante crescente $x[n]=z^n$, para n=0 a 80, parte imaginaria (stem).

```python
n = 80
periodo = 15
frequencia = 1 / periodo
omega = 2 * pi * frequencia
sigma = 0.042

x = arange(0, n+2)

complex_num = exp(complex(sigma, omega) * x*0.7)

complexo_imag = [num.imag for num in complex_num]

plt.stem(x, complexo_imag, use_line_collection=True)
plt.show()
```
